```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
revenue_cleanned=pd.read_csv('MovieRevenue_All_v3.csv',encoding = "ISO-8859-1")
```

```python
#revenue_cleanned['Actors'][1].split(', ')
```

```python
#test Gerne
df_Genre=revenue_cleanned[['Movie_Name','Week_Number','Genre']][revenue_cleanned['Week_Number']==1]
df_Genre[df_Genre['Movie_Name']=='10 Cloverfield Lane']
```

```python
#test Actors
df_Actors=revenue_cleanned[['Movie_Name','Week_Number','Actors']][revenue_cleanned['Week_Number']==1]
df_Actors[df_Actors['Movie_Name']=='10 Cloverfield Lane']
```

```python
df_Genre.info()
```

```python
df=df_Genre
topics=df['Movie_Name'].unique()
for topic in topics:
    #min_week=min(df['Week_Number'][df['Movie_Name']==topic])
    actors_list=df['Genre'][df['Movie_Name']==topic].iloc[0].split(',')
    for i in range(0,len(actors_list)):
        df.loc[-1] = [topic, i-20,actors_list[i]]
        df.index = df.index + 1  # shifting index
        df = df.sort_index()  # sorting by index

df=df.sort_values(by=['Movie_Name','Week_Number'])
df=df[df['Week_Number']!=1]

df=df.reset_index(drop=True)
revenue5_final_5=df
```

```python
df=df_Actors
topics=df['Movie_Name'].unique()
for topic in topics:
    #min_week=min(df['Week_Number'][df['Movie_Name']==topic])
    actors_list=df['Actors'][df['Movie_Name']==topic].iloc[0].split(',')
    for i in range(0,len(actors_list)):
        df.loc[-1] = [topic, i-20,actors_list[i]]
        df.index = df.index + 1  # shifting index
        df = df.sort_index()  # sorting by index

df=df.sort_values(by=['Movie_Name','Week_Number'])
df=df[df['Week_Number']!=1]

df=df.reset_index(drop=True)
revenue5_final_6=df
```

```python
revenue5_final_5.head(20)
```

```python
df_all_Genre=pd.get_dummies(revenue5_final_5, columns=['Genre'],prefix_sep='_',dummy_na=False,drop_first=False)
```

```python
df_all_Actors=pd.get_dummies(revenue5_final_6, columns=['Actors'],prefix_sep='_',dummy_na=False,drop_first=False)
```

```python
df_all_Genre
```

```python
df_all_Actors
```

```python
df=df_all_Genre
topics=df['Movie_Name'].unique()
for topic in topics:
    #min_week=min(df['Week_Number'][df['Movie_Name']==topic])
    #actors_list=df['Genre'][df['Movie_Name']==topic].iloc[0].split(',')
    sumlist=[df_all_Genre[df_all_Genre['Movie_Name']==topic].iloc[0][i] for i in range(2,len(df_all_Genre[df_all_Genre['Movie_Name']==topic].iloc[[0],:].columns))]
    for j in range(1,len(df[df['Movie_Name']==topic])):
        incresae_list=([df_all_Genre[df_all_Genre['Movie_Name']==topic].iloc[j][i] 
                        for i in range(2,len(df_all_Genre[df_all_Genre['Movie_Name']==topic].iloc[[0],:].columns))])
        sumlist=list(map(lambda x, y: x + y, sumlist, incresae_list))
    
    df.loc[-1] = [topic, -30]+sumlist
    df.index = df.index + 1  # shifting index
    df = df.sort_index()  # sorting by index
    
        

df=df.sort_values(by=['Movie_Name','Week_Number'])
df=df[df['Week_Number']==-30]

df=df.reset_index(drop=True)
df_add_Gerne=df
```

```python
df=df_all_Actors
topics=df['Movie_Name'].unique()
for topic in topics:
    #min_week=min(df['Week_Number'][df['Movie_Name']==topic])
    #actors_list=df['Genre'][df['Movie_Name']==topic].iloc[0].split(',')
    sumlist=[df[df['Movie_Name']==topic].iloc[0][i] for i in range(2,len(df[df['Movie_Name']==topic].iloc[[0],:].columns))]
    for j in range(1,len(df[df['Movie_Name']==topic])):
        incresae_list=([df[df['Movie_Name']==topic].iloc[j][i] 
                        for i in range(2,len(df[df['Movie_Name']==topic].iloc[[0],:].columns))])
        sumlist=list(map(lambda x, y: x + y, sumlist, incresae_list))
    
    df.loc[-1] = [topic, -30]+sumlist
    df.index = df.index + 1  # shifting index
    df = df.sort_index()  # sorting by index
    
        

df=df.sort_values(by=['Movie_Name','Week_Number'])
df=df[df['Week_Number']==-30]

df=df.reset_index(drop=True)
df_add_Actors=df
```
